# Task 2 — Time Series Analysis & Decomposition


## Step 1: Import Required Libraries

```python
# Import library for time series decomposition
from statsmodels.tsa.seasonal import seasonal_decompose

# Import Augmented Dickey-Fuller test
from statsmodels.tsa.stattools import adfuller

# Import plotting library
import matplotlib.pyplot as plt
```

## Step 2: Create Monthly Sales Time Series

```python
# Group sales by Order Date
daily_sales = df.groupby("Order Date")["Sales"].sum()

# Convert daily sales to monthly sales
monthly_sales = daily_sales.resample("ME").sum()

# Display first few records
print(monthly_sales.head())
```

## Step 3: Plot Monthly Sales Trend

```python
# Create figure
plt.figure(figsize=(14,6))

# Plot monthly sales
plt.plot(monthly_sales, color="blue")

# Add title and labels
plt.title("Monthly Sales Trend")
plt.xlabel("Order Date")
plt.ylabel("Sales")

# Show graph
plt.show()
```

## Step 4: Apply Time Series Decomposition

```python
# Perform additive decomposition
decomposition = seasonal_decompose(
    monthly_sales,
    model="additive",
    period=12
)
```

## Step 5: Plot All Components

```python
# Plot Trend, Seasonal, Residual and Original series
decomposition.plot()

# Increase figure size
plt.gcf().set_size_inches(15,10)

# Display graph
plt.show()
```

## Step 6: Plot Individual Components

## 1. Trend

```python
# Plot trend component
plt.figure(figsize=(12,4))

plt.plot(decomposition.trend)

plt.title("Trend Component")
plt.show()
```

## 2. Seasonal

```python
# Plot seasonal component
plt.figure(figsize=(12,4))

plt.plot(decomposition.seasonal)

plt.title("Seasonal Component")
plt.show()
```

## 3. Residual

```python
# Plot residual component
plt.figure(figsize=(12,4))

plt.plot(decomposition.resid)

plt.title("Residual Component")
plt.show()
```

## Step 7: Observations
### Observation 1:
#### The sales trend gradually increases over time, indicating business growth.

### Observation 2:
#### A repeating seasonal pattern is visible every year.

### Observation 3:
#### Residual values show random fluctuations that are not explained by trend or seasonality.

### Observation 4:
#### Some months have unusually high or low sales, indicating possible anomalies.

## Step 8: Perform ADF Test

```python
# Perform Augmented Dickey-Fuller Test
result = adfuller(monthly_sales)

# Print ADF Statistic
print("ADF Statistic :", result[0])

# Print p-value
print("p-value :", result[1])

# Print Critical Values
print("Critical Values:")

for key, value in result[4].items():
    print(key, ":", value)
```

## Step 9: Interpret the Result

```python
# Check whether the series is stationary
if result[1] < 0.05:
    print("The time series is Stationary.")
else:
    print("The time series is Non-Stationary.")
```

## Step 10: Apply Differencing (If Non-Stationary)

```python
# Apply first-order differencing
monthly_sales_diff = monthly_sales.diff().dropna()

# Display first few values
print(monthly_sales_diff.head())
```

## Step 11: Plot Differenced Series

```python
# Plot differenced sales
plt.figure(figsize=(14,6))

plt.plot(monthly_sales_diff)

plt.title("Differenced Monthly Sales")

plt.xlabel("Order Date")

plt.ylabel("Differenced Sales")

plt.show()
```

## Step 12: Perform ADF Test Again

```python
# Run ADF test on differenced series
result_diff = adfuller(monthly_sales_diff)

# Print results
print("ADF Statistic :", result_diff[0])
print("p-value :", result_diff[1])

for key, value in result_diff[4].items():
    print(key, ":", value)
```

## Step 13: Final Interpretation

```python
# Interpret the second ADF test
if result_diff[1] < 0.05:
    print("The differenced series is now Stationary.")
else:
    print("The differenced series is still Non-Stationary.")
```

## Step 14: Save Decomposition Charts

```python
# Save decomposition plot
decomposition.plot()
plt.savefig("charts/time_series_decomposition.png")
plt.show()
```

## What is Stationarity?

A stationary time series has constant statistical properties such as mean and variance over time. Forecasting models like SARIMA require stationary data for accurate predictions.

## What Does the ADF Test Tell Us?
p-value < 0.05: The series is stationary.
p-value ≥ 0.05: The series is non-stationary and should be transformed (e.g., by differencing).

## Step 1: Import Required Libraries

```python
# Import SARIMA model
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Import Prophet model
from prophet import Prophet

# Import XGBoost Regressor
from xgboost import XGBRegressor

# Import evaluation metrics
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    mean_absolute_percentage_error
)

# Import NumPy
import numpy as np

# Import Matplotlib
import matplotlib.pyplot as plt
```

## Model 1 – SARIMA

## Step 4: Train SARIMA Model

## Step 1: Import Required Libraries

```python
# Import plotting library
import matplotlib.pyplot as plt

# Import SARIMA model
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Import pandas
import pandas as pd
```

## Step 2: Create Monthly Sales Function

```python
# Function to create monthly sales for a selected column and value
def get_monthly_sales(data, column, value):

    # Filter the data
    filtered = data[data[column] == value]

    # Group by Order Date
    monthly = (
        filtered
        .groupby("Order Date")["Sales"]
        .sum()
        .resample("ME")
        .sum()
    )

    return monthly
```

## Step 3: Create Monthly Sales for Categories

```python
# Furniture sales
furniture = get_monthly_sales(df, "Category", "Furniture")

# Technology sales
technology = get_monthly_sales(df, "Category", "Technology")

# Office Supplies sales
office = get_monthly_sales(df, "Category", "Office Supplies")
```

## Step 4: Create Monthly Sales for Regions

```python
# West region sales
west = get_monthly_sales(df, "Region", "West")

# East region sales
east = get_monthly_sales(df, "Region", "East")
```

## Step 5: Create XGBoost Forecast Function

```python
# Function to forecast the next 3 months using XGBoost
def forecast_sales_xgb(series):

    # Convert series to DataFrame
    data = series.to_frame(name="Sales")

    # Create lag features
    data["Lag1"] = data["Sales"].shift(1)
    data["Lag2"] = data["Sales"].shift(2)
    data["Lag3"] = data["Sales"].shift(3)

    # Create rolling mean feature
    data["RollingMean"] = data["Sales"].rolling(3).mean()

    # Extract month and quarter
    data["Month"] = data.index.month
    data["Quarter"] = data.index.quarter

    # Remove missing values
    data.dropna(inplace=True)

    # Prepare features and target
    X = data.drop("Sales", axis=1)
    y = data["Sales"]

    # Train XGBoost model
    model = XGBRegressor(
        n_estimators=100,
        random_state=42
    )

    model.fit(X, y)

    # Use the last available row as the input for prediction
    last_row = X.iloc[-1:].copy()

    predictions = []

    # Predict the next 3 months
    for _ in range(3):

        # Predict next month's sales
        pred = model.predict(last_row)[0]
        predictions.append(pred)

        # Update lag features
        last_row["Lag3"] = last_row["Lag2"].values
        last_row["Lag2"] = last_row["Lag1"].values
        last_row["Lag1"] = pred

        # Update rolling mean
        last_row["RollingMean"] = (
            last_row["Lag1"] +
            last_row["Lag2"] +
            last_row["Lag3"]
        ) / 3

        # Update month and quarter
        month = int(last_row["Month"].values[0]) + 1
        if month > 12:
            month = 1

        last_row["Month"] = month
        last_row["Quarter"] = ((month - 1) // 3) + 1

    # Create forecast index for the next 3 months
    forecast_index = pd.date_range(
        start=series.index[-1] + pd.offsets.MonthEnd(),
        periods=3,
        freq="ME"
    )

    # Return forecast as a Series
    return pd.Series(predictions, index=forecast_index)
```

## Step 6: Forecast Each Product Category

```python
# Forecast sales for Furniture category
forecast_furniture = forecast_sales_xgb(furniture)

# Forecast sales for Technology category
forecast_technology = forecast_sales_xgb(technology)

# Forecast sales for Office Supplies category
forecast_office = forecast_sales_xgb(office)
```

## Step 7: Forecast Each Region

# Task 5 — Anomaly Detection in Sales Data


## Objective: Detect unusual sales patterns using:

### 1. Isolation Forest
### 2. Z-Score Detection

### Then compare both methods and explain the anomalies.

## Step 1: Import Required Libraries

```python
# Import Isolation Forest model
from sklearn.ensemble import IsolationForest

# Import Z-score function
from scipy.stats import zscore

# Import matplotlib for plotting
import matplotlib.pyplot as plt

# Import pandas
import pandas as pd

# Import NumPy
import numpy as np
```

## Step 2: Prepare Weekly Sales Data

```python
# Group sales by Order Date
daily_sales = df.groupby("Order Date")["Sales"].sum()

# Convert daily sales to weekly sales
weekly_sales = daily_sales.resample("W").sum()

# Convert Series into DataFrame
weekly_df = weekly_sales.to_frame(name="Sales")

# Display first few rows
print(weekly_df.head())
```

## Step 3: Apply Isolation Forest

```python
# Create Isolation Forest model
iso_model = IsolationForest(
    contamination=0.05,
    random_state=42
)

# Train the model
weekly_df["Isolation_Anomaly"] = iso_model.fit_predict(
    weekly_df[["Sales"]]
)
```

## Step 4: Convert Predictions into Labels

```python
# Convert predictions into readable labels
weekly_df["Isolation_Label"] = weekly_df[
    "Isolation_Anomaly"
].map({
    1: "Normal",
    -1: "Anomaly"
})

# Display detected anomalies
print(
    weekly_df[
        weekly_df["Isolation_Label"] == "Anomaly"
    ]
)
```

## Step 5: Plot Isolation Forest Anomalies:

```python
# Create figure
plt.figure(figsize=(14,6))

# Plot weekly sales
plt.plot(
    weekly_df.index,
    weekly_df["Sales"],
    label="Weekly Sales"
)

# Highlight anomalies
plt.scatter(
    weekly_df[
        weekly_df["Isolation_Label"]=="Anomaly"
    ].index,

    weekly_df[
        weekly_df["Isolation_Label"]=="Anomaly"
    ]["Sales"],

    color="red",

    label="Anomaly"
)

# Add title and labels
plt.title("Isolation Forest Anomaly Detection")

plt.xlabel("Date")

plt.ylabel("Sales")

plt.legend()

plt.show()
```

## Step 6: Apply Z-Score Detection

```python
# Calculate Z-score
weekly_df["Z_Score"] = zscore(
    weekly_df["Sales"]
)

# Mark anomalies
weekly_df["Z_Anomaly"] = np.where(
    abs(weekly_df["Z_Score"]) > 2,
    "Anomaly",
    "Normal"
)

# Display anomalies
print(
    weekly_df[
        weekly_df["Z_Anomaly"]=="Anomaly"
    ]
)
```

## Step 7: Plot Z-Score Anomalies

```python
# Create figure
plt.figure(figsize=(14,6))

# Plot weekly sales
plt.plot(
    weekly_df.index,
    weekly_df["Sales"],
    label="Weekly Sales"
)

# Highlight anomalies
plt.scatter(
    weekly_df[
        weekly_df["Z_Anomaly"]=="Anomaly"
    ].index,

    weekly_df[
        weekly_df["Z_Anomaly"]=="Anomaly"
    ]["Sales"],

    color="orange",

    label="Z-Score Anomaly"
)

# Add title and labels
plt.title("Z-Score Anomaly Detection")

plt.xlabel("Date")

plt.ylabel("Sales")

plt.legend()

plt.show()
```

## Step 8: Compare Both Methods

```python
# Create comparison table
comparison = weekly_df[[
    "Sales",
    "Isolation_Label",
    "Z_Anomaly"
]]

# Display comparison
print(comparison)
```

# Task 6 — Product Demand Segmentation using Clustering

## Objective: Segment products based on demand patterns using K-Means Clustering and visualize the clusters using PCA.

## Step 1: Import Required Libraries

```python
# Import pandas
import pandas as pd

# Import NumPy
import numpy as np

# Import K-Means clustering
from sklearn.cluster import KMeans

# Import StandardScaler for feature scaling
from sklearn.preprocessing import StandardScaler

# Import PCA for dimensionality reduction
from sklearn.decomposition import PCA

# Import matplotlib
import matplotlib.pyplot as plt

# Import seaborn
import seaborn as sns
```

## Step 2: Aggregate Product-Level Features

```python
product_data = df.groupby("Sub-Category").agg({

    "Sales": ["sum", "mean", "std"]

})

# Rename columns
product_data.columns = [

    "Total_Sales",
    "Average_Order_Value",
    "Sales_Volatility"

]

# Replace missing values
product_data.fillna(0, inplace=True)

# Display dataset
print(product_data.head())
```

## Step 3: Calculate Year-over-Year Growth Rate

```python
# Calculate yearly sales by sub-category
yearly_sales = df.groupby(
    ["Sub-Category", "Year"]
)["Sales"].sum().unstack(fill_value=0)

# Calculate growth rate
growth_rate = (
    yearly_sales.iloc[:, -1] -
    yearly_sales.iloc[:, 0]
) / yearly_sales.iloc[:, 0]

# Replace infinite values
growth_rate = growth_rate.replace(
    [np.inf, -np.inf],
    0
).fillna(0)

# Add growth rate to dataset
product_data["Growth_Rate"] = growth_rate
```

## Step 4: Select Features

```python
# Select features for clustering
features = product_data[

    [

        "Total_Sales",
        "Growth_Rate",
        "Sales_Volatility",
        "Average_Order_Value"

    ]

]
```

## Step 5: Standardize Features

```python
# Create scaler
scaler = StandardScaler()

# Scale features
scaled_features = scaler.fit_transform(features)
```

## Step 6: Find Optimal Number of Clusters (Elbow Method)

```python
# Store inertia values
inertia = []

# Test different values of K
for k in range(1,11):

    model = KMeans(
        n_clusters=k,
        random_state=42
    )

    model.fit(scaled_features)

    inertia.append(model.inertia_)
```

## Step 7: Plot Elbow Curve

```python
# Create figure
plt.figure(figsize=(8,5))

# Plot elbow curve
plt.plot(

    range(1,11),

    inertia,

    marker="o"

)

# Add labels
plt.title("Elbow Method")

plt.xlabel("Number of Clusters")

plt.ylabel("Inertia")

plt.show()
```

## Step 8: Apply K-Means Clustering

```python
# Create KMeans model
kmeans = KMeans(
    n_clusters=4,
    random_state=42
)

# Predict clusters
product_data["Cluster"] = kmeans.fit_predict(
    scaled_features
)
```

## Step 9: Reduce Dimensions using PCA

```python
# Create PCA model
pca = PCA(n_components=2)

# Reduce dimensions
pca_features = pca.fit_transform(
    scaled_features
)

# Store PCA values
product_data["PCA1"] = pca_features[:,0]
product_data["PCA2"] = pca_features[:,1]
```

## Step 10: Plot Product Clusters

```python
# Create figure
plt.figure(figsize=(10,6))

# Plot clusters
sns.scatterplot(

    data=product_data,

    x="PCA1",

    y="PCA2",

    hue="Cluster",

    s=120

)

# Add title
plt.title("Product Demand Segments")

plt.show()
```

## Step 11: Assign Cluster Labels

```python
# Assign business labels
cluster_names = {

    0: "High Volume, Stable Demand",

    1: "Low Volume, High Volatility",

    2: "Growing Demand",

    3: "Declining Demand"

}

# Create Demand Segment column
product_data["Demand_Segment"] = product_data[
    "Cluster"
].map(cluster_names)

# Display results
print(product_data[
    [
        "Demand_Segment"
    ]
])
```

## Step 12: Display Final Segmentation Table

```python
# Display important columns
print(product_data[

    [

        "Total_Sales",

        "Growth_Rate",

        "Sales_Volatility",

        "Average_Order_Value",

        "Demand_Segment"

    ]

])
```

```python
# Save chart
plt.savefig(
    "charts/product_clusters.png"
)
```
